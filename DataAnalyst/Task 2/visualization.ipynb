### **Preprocessing**

#### **importing necessary packages**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### ***Loading Dataset***

```python
iris_data=pd.read_csv("iris.csv")
```

```python
iris_data.head()
```

#### **Removing Unwanted columns**

```python
iris_data.duplicated().sum()
```

```python
#romoving id attribute because it's not important for classification
iris_data.drop("Id",axis=1,inplace=True)
```

```python
iris_data.head()
```

### **Visualization**

#### **visualize SepalLength column**

```python
plt.hist(iris_data['SepalLengthCm'],color='blue')
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Iris data Sepal Length")
plt.show()
```

#### **visualize SepalWidth column**

```python
plt.hist(iris_data['SepalWidthCm'],color='red',alpha=0.7)
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.title("Histogram of Iris data Sepal Width")
plt.show()
```

#### **visualize PetalLenth column**

```python
plt.hist(iris_data['PetalLengthCm'],color='orange')
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Iris data Petal Length")
plt.show()
```

#### **visualize PetalWidth column**

```python
plt.hist(iris_data['PetalWidthCm'],color='yellow')
plt.xlabel("Petal Width")
plt.ylabel("Frequency")
plt.title("Histogram of Iris data Petal Width")
plt.show()
```

#### **visualize Species column**

```python
plt.hist(iris_data['Species'],color='navy')
plt.xlabel("Species")
plt.ylabel("Frequency")
plt.title("Histogram of Iris data Species")
plt.show()
```
